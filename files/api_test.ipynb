# Тестирование API запросов

`В данном документе представлено получение данных от приложения по API запросам`

```python
import requests
from datetime import date

BASE_URL = "https://api.vkr-klauzer.ru"

def test_health():
    r = requests.get(f"{BASE_URL}/health")
    print("HEALTH:", r.status_code, r.json())


def test_pairs():
    r = requests.get(f"{BASE_URL}/pairs")
    print("PAIRS:", r.status_code, r.json())


def test_history():
    params = {
        "start": "2024-01-01",
        "end": "2025-01-01"
    }
    pair = "EUR/USD"  # замени на реальный ключ из PAIRS
    r = requests.get(f"{BASE_URL}/history/{pair}", params=params)
    print("HISTORY:", r.status_code, r.json()[:2])


def test_context():
    params = {
        "start": "2024-01-01",
        "end": "2025-01-01"
    }
    pair = "EUR/USD"
    r = requests.get(f"{BASE_URL}/context/{pair}", params=params)
    print("CONTEXT:", r.status_code, r.json()[:2])


def test_predictions():
    params = {
        "start": "2024-01-01",
        "end": "2025-01-01",
        "limit": 10
    }
    pair = "EUR/USD"
    r = requests.get(f"{BASE_URL}/predictions/{pair}", params=params)
    print("PREDICTIONS:", r.status_code, r.json()[:2])


def test_latest():
    pair = "EUR/USD"
    r = requests.get(f"{BASE_URL}/latest/{pair}")
    print("LATEST:", r.status_code, r.json())


def test_etl_stats():
    r = requests.get(f"{BASE_URL}/etl/stats", params={"days": 30})
    print("ETL STATS:", r.status_code, r.json())
```
